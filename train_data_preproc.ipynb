```python
import numpy as np
import pandas as pd


def symifactor():
    
    alfa = np.exp(1j * 2 * np.pi/3)
    sym_factor = np.array([
        [1, 1, 1],
        [1, alfa**2, alfa],
        [1, alfa, alfa**2]
    ])
    return sym_factor
```
