```python
import numpy as np
import pandas as pd
deltaTime = 0.01
time = np.arange(0, 41*deltaTime, deltaTime)
G_1 = pd.DataFrame(
    {
        'time': time,
        'current A beg': clf.current_A_begin['exp 1'],
        'current A end': clf.current_A_end['exp 1'],
        'current B beg': clf.current_B_begin['exp 1'],
        'current B end': clf.current_B_end['exp 1'],
        'current C beg': clf.current_C_begin['exp 1'],
        'current C end': clf.current_C_end['exp 1'],
        'current A angle begin': clf.current_A_angle_begin['exp 1'],
        'current A angle end': clf.current_A_angle_end['exp 1'],
        'current B angle begin': clf.current_B_angle_begin['exp 1'],
        'current B angle end': clf.current_B_angle_end['exp 1'],
        'current C angle begin': clf.current_C_angle_begin['exp 1'],
        'current C angle end': clf.current_C_angle_end['exp 1'],
        'voltage A begin': clf.voltage_A_begin['exp 1'],
        'voltage A end': clf.voltage_A_end['exp 1'],
        'voltage B begin': clf.voltage_B_begin['exp 1'],
        'voltage B end': clf.voltage_B_end['exp 1'],
        'voltage C begin': clf.voltage_C_begin['exp 1'],
        'voltage C end': clf.voltage_C_end['exp 1'],
        'voltage A angle begin': clf.voltage_A_angle_begin['exp 1'],
        'voltage A angle end': clf.voltage_A_angle_end['exp 1'],
        'voltage B angle begin': clf.voltage_B_angle_begin['exp 1'],
        'voltage B angle end': clf.voltage_B_angle_end['exp 1'],
        'voltage C angle begin': clf.voltage_C_angle_begin['exp 1'],
        'voltage C angle end': clf.voltage_C_angle_end['exp 1']
    }
    )
G_1 = G_1.set_index('time')
```

```python
from matplotlib import pyplot as plt

axes_names = [
    ", A",
    " , .",
    ", ",
    " , ."
]
line_name_I = [
    r'$I_{beg}^{A}$',
    r'$I_{end}^{A}$',
    r'$I_{beg}^{B}$',
    r'$I_{end}^{B}$',
    r'$I_{beg}^{C}$',
    r'$I_{end}^{C}$'
    ]
line_name_I_ang = [
    r'$\varphi_{I beg}^{A}$',
    r'$\varphi_{I end}^{A}$',
    r'$\varphi_{I beg}^{B}$',
    r'$\varphi_{I end}^{B}$',
    r'$\varphi_{I beg}^{C}$',
    r'$\varphi_{I end}^{C}$'
    ]
line_name_U = [
    r'$U_{beg}^{A}$',
    r'$U_{end}^{A}$',
    r'$U_{beg}^{B}$',
    r'$U_{end}^{B}$',
    r'$U_{beg}^{C}$',
    r'$U_{end}^{C}$']
line_name_U_ang = [
    r'$\varphi_{U beg}^{A}$',
    r'$\varphi_{U end}^{A}$',
    r'$\varphi_{U beg}^{B}$',
    r'$\varphi_{U end}^{B}$',
    r'$\varphi_{U beg}^{C}$',
    r'$\varphi_{U end}^{C}$'
    ]

line_style = [
    '-',
    '--',
    '-.',
    '--',
    'dotted',
    '--'
]
idx_1 = 0
idx_2 = 0
idx_3 = 0
idx_4 = 0



fig, axes = plt.subplots(2, 2, facecolor='white', figsize=(15, 10))


for col in G_1.columns:
    if 'current' in col.split() and 'angle' not in col.split():
        G_1[col].plot(ax=axes[0,0], linewidth=3, legend=True, linestyle=line_style[idx_1], label=line_name_I[idx_1])

        idx_1+=1

    elif 'current' in col.split() and 'angle' in col.split():
        G_1[col].plot(ax=axes[0,1], linewidth=3, legend=True, linestyle=line_style[idx_2], label=line_name_I_ang[idx_2])
        axes[0,1].set_xlabel(', ', fontsize=12)
        axes[0,1].set_ylabel(axes_names[1], fontsize=12)
        axes[0,1].tick_params(axis='both', labelsize=12)
        axes[0,1].grid(True)
        axes[0,1].bacgroundcolor='white'
        axes[0,1].linecolor = 'black'
        axes[0,1].grid(True, color='grey', linestyle='-', linewidth=1)
        axes[0,1].legend(fontsize=14, loc='lower left', shadow=True)
        leg2=axes[0,1].legend(fontsize=14, shadow=True)
        leg2.get_frame().set_facecolor('white')
        axes[0,1].set_facecolor('white')
        idx_2+=1
    elif 'voltage' in col.split() and 'angle' not in col.split():
        G_1[col].plot(ax=axes[1,0], linewidth=3, legend=True, linestyle=line_style[idx_3], label=line_name_U[idx_3])
        axes[1,0].set_xlabel(', ', fontsize=12)
        axes[1,0].set_ylabel(axes_names[2], fontsize=12)
        axes[1,0].tick_params(axis='both', labelsize=12)
        axes[1,0].grid(True)
        axes[1,0].bacgroundcolor='white'
        axes[1,0].linecolor = 'black'
        axes[1,0].grid(True, color='grey', linestyle='-', linewidth=1)
        leg3=axes[1,0].legend(fontsize=14, shadow=True)
        leg3.get_frame().set_facecolor('white')
        axes[1,0].set_facecolor('white')
        idx_3+=1
    elif 'voltage' in col.split() and 'angle' in col.split():
        G_1[col].plot(ax=axes[1,1], linewidth=3, legend=True, linestyle=line_style[idx_4], label=line_name_U_ang[idx_4])
        axes[1,1].set_xlabel(', ', fontsize=12)
        axes[1,1].set_ylabel(axes_names[3], fontsize=12)
        axes[1,1].tick_params(axis='both', labelsize=12)
        axes[1,1].grid(True)
        axes[1,1].bacgroundcolor='white'
        axes[1,1].linecolor = 'black'
        axes[1,1].grid(True, color='grey', linestyle='-', linewidth=1)
        leg4=axes[1,1].legend(fontsize=14, loc='lower left', shadow=True)
        leg4.get_frame().set_facecolor('white')
        axes[1,1].set_facecolor('white')
        idx_4+=1

axes[0,0].set_xlabel(', ', fontsize=12)
axes[0,0].set_ylabel(axes_names[0], fontsize=12)
axes[0,0].tick_params(axis='both', labelsize=12)
axes[0,0].grid(True)
axes[0,0].bacgroundcolor='white'
axes[0,0].linecolor = 'black'
axes[0,0].grid(True, color='grey', linestyle='-', linewidth=1)
leg1=axes[0,0].legend(fontsize=14, shadow=True)
leg1.get_frame().set_facecolor('white')
axes[0,0].set_facecolor('white')

plt.tight_layout()
plt.show()
```
