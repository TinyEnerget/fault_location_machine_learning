```python
def rename_columns(data_current):
    columns_name = []
    for indx in range(len(data_current.columns)):
        columns_name.append('feature ' + str(indx + 1))

    data_current.columns = columns_name
    return data_current
```

```python
# Выделение методов

file_path = "delta method/"
file_name = 'delta_per_method.csv'
delta_method = di(';', file_path + file_name).main_process()

best_method = pd.DataFrame()
for indx in range(len(delta_method.columns)):
    exp_num = "exp " + str(indx + 1)
    tmp = pd.DataFrame({
        exp_num: [abs(delta_method[exp_num]).idxmin() + 1]
        })
    best_method = pd.concat([best_method, tmp], axis=1)

best_method.to_csv('best_method.csv', sep=';', index=False)
```
